```python
bins = [0, 1/3, 2/3, 1]
labels=['Low', 'Medium', 'High']
# Create bins for each attribute
df['danceability_bin'] = pd.qcut(df['danceability'], bins, labels)
df['energy_bin'] = pd.qcut(df['energy'], bins, labels)
df['speechiness_bin'] = pd.qcut(df['speechiness'], bins, labels)
df['acousticness_bin'] = pd.qcut(df['acousticness'], bins, labels)
df['instrumentalness_bin'] = pd.qcut(df['instrumentalness'], bins, labels)
df['liveness_bin'] = pd.qcut(df['liveness'], bins, labels)
df['valence_bin'] = pd.qcut(df['valence'], bins, labels)
```

```python
df['loudness_bin'] = pd.cut(df['loudness'], bins=[-float('inf'), -10, -5, float('inf')],
                                  labels=['Low', 'Medium', 'High'], include_lowest=True)
df['duration_bin'] = pd.cut(df['duration_ms'], bins=[-float('inf'), 180000, 240000, float('inf')], labels=['Short', 'Medium', 'Long'], include_lowest=True)
```

```python
df['key_new'] = df['key'].astype(str)
```

```python
df['tempo_bin'] = pd.cut(df['tempo'], bins=[-float('inf'), 100, 130, float('inf')], labels=['Slow', 'Medium', 'Fast'], include_lowest=True)
df['popularity_bin']=pd.cut(df['popularity'], bins=[-float('inf'), 50, 75, float('inf')],labels=['Low', 'Medium', 'High'], include_lowest=True)  
mode_dict = {0: "Minor", 1: "Major"}
df["mode_bin"] = df["mode"].map(mode_dict)
```

```python
df['danceability_bin']  = df['danceability_bin'].astype(str) + "_danceability"
df['energy_bin']  = df['energy_bin'] .astype(str) + "_energy"
df['speechiness_bin']  = df['speechiness_bin'].astype(str) + "_speechniess"
df['acousticness_bin']  = df['acousticness_bin'].astype(str) + "_acousticness"
df['instrumentalness_bin']  = df['instrumentalness_bin'].astype(str) + "_instrumentalness"
df['liveness_bin']  = df['liveness_bin'].astype(str) + "_liveness"
df['valence_bin']  = df['valence_bin'].astype(str) + "_valence"
df['loudness_bin']  = df['loudness_bin'].astype(str) + "_loudness"
df['key_new']= df['key_new'].astype(str) + "_key"
df['tempo_bin']= df['tempo_bin'].astype(str) + "_tempo"
df['tduration_bin']= df['duration_bin'].astype(str) + "_duration"
df['popularity_bin']= df['popularity_bin'].astype(str) + "_popularity"
df['mode_bin']= df['mode_bin'].astype(str)
df['duration_bin']= df['duration_bin'].astype(str)+ "_duration"
```

### Support plots

#### Support vs number of itemsets

```python
import matplotlib.pyplot as plt
import numpy as np
len_max_it = []
len_cl_it = []
max_supp = 25
for i in range(2, max_supp):
    max_itemsets = apriori(X, target="m", supp=i, zmin=3)
    cl_itemsets = apriori(X, target="c", supp=i, zmin=3)
    len_max_it.append(len(max_itemsets))
    len_cl_it.append(len(cl_itemsets))

plt.plot(np.arange(2, max_supp), len_max_it, label="maximal")
plt.plot(np.arange(2, max_supp), len_cl_it, label="closed")
plt.legend()
plt.xlabel("%support")
plt.ylabel("itemsets")

plt.show()
```

### Support vs number of itemset for Major / Minor

```python
filter_1 = []
filter_2 = []
filter_3 = []
for i in range(2, max_supp):
    max_itemsets = apriori(X, target="m", supp=i, zmin=zmin)
    filter_1.append(len([item for item in max_itemsets if "Major" in item[0]]))
    filter_2.append(len([item for item in max_itemsets if "Minor" in item[0]]))
plt.plot(np.arange(2, max_supp), filter_1, label="Major")
plt.plot(np.arange(2, max_supp), filter_2, label="Minor")
plt.legend()
plt.xticks()
plt.xlabel("%support")
plt.ylabel("itemsets")

plt.show()
```

```python
def plot_frequent_patterns(target, title):
    res = {}
    for i in np.arange(10, 26, 1):
        supp = round(i,2)
        sets = apriori(X, supp=supp, zmin=3, target=target)
        res[supp] = len(sets)

    plt.bar(range(len(res)), list(res.values()), align='center')
    plt.xticks(range(len(res)), list(res.keys()))


    plt.xlabel("Support")
    plt.ylabel("Number of Itemsets")
    plt.title("{} itemsets".format(title))
    plt.show()

plot_frequent_patterns('s', "Frequent itemsets")
plot_frequent_patterns('c', "Closed Frequent itemsets")
plot_frequent_patterns('m', "Maximal Frequent itemsets")
```

### Association Rule

```python
conf = 60
rules = apriori(X, target="r", supp=supp, zmin=3, conf=conf, report="aScl")
rules_df = pd.DataFrame(
    rules,
    columns=[
        "consequent",
        "antecedent",
        "abs_support",
        "%_support",
        "confidence",
        "lift",
    ],
)
rules_df[:10].sort_values(by="lift", axis=0, ascending=False)
```

```python
def plot_rules_confidence():
    bar_res = {}
    for i in np.arange(10, 100, 10):
        rls = apriori(X, supp=20, zmin=3, conf=i, target='r')
        bar_res[i] = len(rls)

    plt.bar(range(len(bar_res)), list(bar_res.values()), align='center')
    plt.xticks(range(len(bar_res)), list(bar_res.keys()))

    plt.xlabel("Confidence")
    plt.ylabel("Number of Rules")
    plt.show()

plot_rules_confidence()
```

```python
def plot_conf_support():
    res = []
    for supp in np.arange(1, 21, 1):
        for confidence in np.arange(10, 100,10):
            support = round(supp,2)
            rls = apriori(X, supp=support, conf=confidence, zmin=3, target='r', report='SCl')
            for r in rls:
                res.append({ 'supp': r[2], 'conf': r[3], 'lift': r[4]})

    df_results = pd.DataFrame(res)
    df_results = df_results.sample(10000)

    plt.scatter(df_results['conf'], df_results['supp'], c=df_results['lift'], cmap='viridis')
    plt.xlabel("Confidence")
    plt.ylabel("Support")
    plt.colorbar(label="Lift")
    plt.show()
    print(df_results)
plot_conf_support()
```

## Predict the target variable

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
```

```python
# Use Mode_bin column with value "Major" as the target, with the input includes: "Slow_tempo" and "Low_popularity"
X_train, X_test, y_train, y_test = train_test_split(
    df.drop("mode_bin", axis=1),
    df["mode_bin"],
    test_size=0.2, 
    random_state=0,
    stratify=df["mode_bin"]
)
```

```python
y_pred = X_test.apply(lambda x: "Major" if (x["tempo_bin"] == "Slow_tempo" and x["popularity_bin"] == "Low_popularity") else "Minor", axis=1)
```

```python
cm = confusion_matrix(y_test, y_pred, labels=["Major", "Minor"])
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Major", "Minor"])
```

```python
disp.plot(cmap='viridis')
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred, target_names=["Major", "Minor"])
print(report)
```
